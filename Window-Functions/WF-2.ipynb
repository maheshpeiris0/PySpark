```python
from pyspark.sql import functions as F
```

```python
dataset =df.groupBy('Ticker').agg(F.sum('day_open').alias('sum_day_open')).orderBy('sum_day_open',ascending=True)
```

```python
dataset.filter(dataset['sum_day_open']==0).count()
```

```python
drop_df=dataset.filter(dataset['sum_day_open']==0)
```

```python
df.select('Ticker').distinct().count()
```

```python
drop_df.select('Ticker').distinct().count()
```

```python
new_df = df.join(drop_df, df['Ticker']==drop_df['Ticker'],'left_anti')
```

```python
new_df.select('Ticker').distinct().count()  
```

```python
new_df.filter(new_df['Ticker']=='AAPL').show(20)
```

```python
new_df.printSchema()
```

```python
from pyspark.sql.window import Window
```

```python
windowsSpec = Window.partitionBy(new_df['Ticker']).orderBy(new_df['datetime_est'].desc())
```

```python
new_df=new_df.withColumn('row_number',F.row_number().over(windowsSpec))
```

```python
max_time_df = new_df.filter(new_df['row_number']==1).drop('row_number')
```
